```python
import os 
from pytubefix import YouTube
from pytubefix.exceptions import VideoUnavailable, VideoPrivate
import numpy as np 
import cv2
import videokf as vf
```

```python
train_data_file = "./dataset/train_videoxum.json" # change path according where the file is stored 
with open(train_data_file, 'r') as file:
    train_data = json.load(file)
```

```python
youtube_data = {
    "url": [], 
    "file": [] 
} 

for i in range(len(train_data[:10])): 
    datapoint = train_data[i] 
    video_id = datapoint["video_id"] 
    video_file = video_id.split("_")[-1]
    youtube_file = "https://www.youtube.com/watch?v="+video_file

    file_dir = "./dataset/video_folder/" 
    if not os.path.exists("./dataset"): 
        os.mkdir("./dataset") 
    if not os.path.exists(file_dir): 
        os.mkdir(file_dir)
        
    file_path = os.path.join(file_dir, video_file+".mp4")
    print(youtube_file)
    try: 
        yt = YouTube(youtube_file) 
        stream = yt.streams.get_highest_resolution()
        stream.download(output_path="./dataset/video_folder/", filename=video_file+".mp4")
        youtube_data["url"].append(youtube_file) 
        youtube_data["file"].append(video_file)
    except VideoUnavailable:
        print("The video is unavailable.")
    except VideoPrivate:
        print("The video is private.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
if not os.path.exists("./dataset/keyframes_videokf"): 
    os.mkdir("./dataset/keyframes_videokf")
    
for video_file in youtube_data["file"]:
    file_dir = "./dataset/video_folder/"     
    file_path = os.path.join(file_dir, video_file+".mp4")

    keyframe_folder = os.path.join("./dataset/keyframes_videokf", video_file)
    if not os.path.exists(keyframe_folder): 
        os.mkdir(keyframe_folder)
        vf.extract_keyframes(file_path, method="iframes",  output_dir_keyframes=keyframe_folder)
```

```python
# first method of image captioning 
from IPython.display import display
import requests
import torch
from PIL import Image
from transformers import *
from tqdm import tqdm
import urllib.parse as parse
import os

device = "cuda" if torch.cuda.is_available() else "cpu"
finetuned_model = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning").to(device)
finetuned_tokenizer = GPT2TokenizerFast.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
finetuned_image_processor = ViTImageProcessor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")

def get_caption(model, image_processor, tokenizer, image_path):
    image = Image.open(image_path)
    img = image_processor(image, return_tensors="pt").to(device)
    output = model.generate(**img)
    caption = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
    return caption


img_file = "./dataset/keyframes_videokf/fzp5ooc727c/1988.jpg"
display(img_file)
caption = get_caption(finetuned_model, finetuned_image_processor, finetuned_tokenizer, img_file)
image = Image.open(img_file)
image.show()
print(caption)
```
