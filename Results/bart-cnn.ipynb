```python
import numpy as np
import pandas as pd
import torch
from transformers import *
from tqdm import tqdm
import os
from transformers import BartTokenizer
from torch import cuda
device = 'cuda' if cuda.is_available() else 'cpu'

video_caption_dict = np.load('/home1/kmayya/Pipeline/Pipeline/video_caption_dict.npy',allow_pickle='TRUE').item()

# tokenizer = BartTokenizer.from_pretrained("facebook/bart-large")
# from transformers import pipeline
# model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
# model.to(device)
# summarizer = pipeline("summarization", model=model, tokenizer=tokenizer, device=device)
# summary_dict = {} 

# def split_text_into_chunks(text, max_length=1024):
#     tokens = tokenizer(text, return_tensors="pt")['input_ids'][0]
#     chunks = [tokens[i:i + max_length] for i in range(0, len(tokens), max_length)]
#     return [tokenizer.decode(chunk, skip_special_tokens=True) for chunk in chunks]

# for key, val in video_caption_dict.items():
#     if key == '.ipynb_checkpoints':
#         continue 
#     print("video = ", key) 
#     document = " ".join(val) 

#     chunks = split_text_into_chunks(document, max_length=1024)
#     chunk_summaries = []
#     #for chunk in chunks:
#     final_summary =  summarizer(chunk[0], max_length=200, min_length=150, do_sample=False)
#     #chunk_summaries.append(summary[0]['summary_text'])
    
#     #final_summary = " ".join(chunk_summaries)
#     print("Generated Paragraph:", final_summary)
#     summary_dict[key] = final_summary
from transformers import BartTokenizer

tokenizer = BartTokenizer.from_pretrained("facebook/bart-large")
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", tokenizer=tokenizer, device=device)
summary_dict = {} 

for i, (key, val) in enumerate(video_caption_dict.items()):
    if i >= 1000:
        break
    print(f"video ${i} = ", key) 
    document = " ".join(val) 
    tokens = tokenizer(document, max_length=1024, truncation=True, return_tensors="pt")
    truncated_text = tokenizer.decode(tokens['input_ids'][0], skip_special_tokens=True)
    summary = summarizer(truncated_text, max_length=200, min_length=150, do_sample=False)
    
    summary_output = summary[0]['summary_text']
    # print("Generated Paragraph:", summary_output)
    summary_dict[key] = summary_output
```

```python
df = pd.read_csv('summarized_dense_descriptions_1000_final.csv')
df['bart_summarized_video_captions'] = df['video_id'].map(summary_dict)
df.to_csv('/home1/kmayya/Pipeline/summarized_dense_descriptions_1000_final.csv', index=False) 
```
